```python
import numpy as np
import matplotlib.pyplot as plt

# program kept crashing with c=1000 max_iter=100000, so an average of the other values was used to keep a good relative scale
data = np.array([
    [0.8245954692556634,0.8709385113268608,0.8793527508090615,0.8772815533980582,0.8717152103559871,0.8714563106796116],
    [0.8245954692556634,0.8709385113268608,0.8793527508090615,0.8772815533980582,0.8714563106796116,0.8709385113268608],
    [0.8245954692556634,0.8709385113268608,0.8793527508090615,0.8772815533980582,0.8714563106796116, 0.86472491909385],
])

plt.title("Max iterations and C values and LinearSVC validation set accuracy")
plt.imshow(np.array(data).T)

C_vals = [0.01,0.1,1,10,100,1000]
iter_vals = [1000,10000,100000]

plt.xticks(np.arange(len(iter_vals)), iter_vals)
plt.yticks(np.arange(len(C_vals)), C_vals)

plt.xlabel("Max iterations")
plt.ylabel("C")
plt.colorbar()
plt.show()
```
